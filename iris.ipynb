```python
from sklearn import datasets
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

# load iris dataset
dataset = datasets.load_iris()
data = pd.DataFrame(dataset.data,columns=["Petal Length","Petal Width","Sepal Length","Sepal Width"])
data['Species'] = dataset.target
data.head()
```

```python
# PCA/LDA dimension reduction
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = data[["Petal Length","Petal Width","Sepal Length","Sepal Width"]]
X = StandardScaler().fit_transform(X)
Y = dataset.target
target_values = sorted(list(set(Y)))
target_names = dataset.target_names

PCA = PCA(n_components=2)
X_pca = PCA.fit_transform(X, Y)

LDA = LinearDiscriminantAnalysis(n_components=2)
X_lda = LDA.fit_transform(X, Y)

X_pca_lda = LDA.fit_transform(X_pca, Y)

def subplot_scatter(X, Y, target_values, target_names, fig, subplot_location=None, title=None, set_legend=True):
    ax=fig.add_subplot(subplot_location)
    
    # create a scatter plot parsing through data points of the matrix, and their corresponding labeled outputs
    for i, target in zip(target_values, target_names):
        ax.scatter(X[Y == i, 0], X[Y == i, 1], label=target)
    
    if set_legend == True:  # add a legend if set_legend is True
        ax.legend(loc='best',scatterpoints=1)
        
    if title: # add a title if not null
        ax.set_title(title)
        
fig=plt.figure(figsize=(12,10))
subplot_scatter(X_pca, Y, target_values, target_names, fig, 131, "PCA of IRIS dataset", True)
subplot_scatter(X_lda, Y, target_values, target_names, fig, 132, "LDA of IRIS dataset", True)
subplot_scatter(X_pca_lda, Y, target_values, target_names, fig, 133, "PCA-LDA of IRIS dataset", True)

plt.show()
```

```python
# CLASSIFIER - Evaluation of the techniques via confusion matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score

classifier = RandomForestClassifier(max_depth=2, random_state=42)

def perform_evaluation(feature_set,target_set):
    def namestr(obj, namespace):
        return [name for name in namespace if namespace[name] is obj]
    print("Evaluation for:",namestr(feature_set,globals())[0])
    feat_train, feat_test, target_train, target_test = train_test_split(feature_set, target_set, test_size=0.2, random_state=42)

    classifier.fit(feat_train, target_train)
    predicted = classifier.predict(feat_test)
    print("Accuracy: ",accuracy_score(target_test, predicted))
    print(confusion_matrix(target_test, predicted))
    
perform_evaluation(X, Y)
perform_evaluation(X_pca, Y)
perform_evaluation(X_lda, Y)
perform_evaluation(X_pca_lda, Y)
```
